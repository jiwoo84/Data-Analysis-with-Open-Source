**로또 추첨 프로그램**

1 ~ 45 숫자 6개를 입력 받아 당첨 숫자와 비교

```python
import random
guess_str = input("1~45 번호 6개를 입력하세요: ").split(", ")

guess_list = [int(i) for i in guess_str if int(i) < 46]

if len(guess_list) == 6:
    lotto_list = random.sample(range(1, 46, 1), 6) # sample:중복되지 않게 여러개의 원소 선택
    print("당첨 숫자는", lotto_list, "입니다.")
    hit_count = 0

    for i in guess_list:
        for j in lotto_list:
            if i == j:
                hit_count += 1

    print(str(hit_count) + "개 맞혔습니다.")
else :
    print("45 이하의 번호 6개를 넣으세요")
```

**스무고개 프로그램**

20번의 기회 안에 1 ~ 100 사이의 숫자를 맞히기


```python
import random
hit_number = random.randint(1, 101)
# print(hit_number)

guess_count_list = range(1, 21, 1)
passed = False
for guess_count in guess_count_list :
    guess = int(input("숫자를 맞춰보세요(" + str(guess_count) + "번째 시도): "))

    if hit_number == guess:
        passed = True
        break
    elif hit_number > guess:
        print(str(guess) + "보다 큽니다.", end="")
    else:
        print(str(guess) + '보다 작습니다', end="")

if passed == True:
    print("맞혔습니다. 축하합니다.")
else:
    print("모든 기회를 다 사용하셨습니다. ")
```

**5000보다 작은 소수 찾기**

```python
import time

max = 50000
def Isprime(x):
    for i in range(2, x):
        if x % i == 0:
            return False
        return True

pn_count = 0
start_time = time.time()
for i in range(2, max+1):
    if Isprime(i):
        pn_count += 1
        print(i, end=", ")
        if(pn_count % 5 == 0):
            print("")

print("")
end_time = time.time()

print(format(end_time - start_time, ".3f"), "초 실행습니다." )
```
