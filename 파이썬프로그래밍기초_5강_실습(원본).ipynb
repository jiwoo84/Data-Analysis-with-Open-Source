**원뿔의 부피&겉넓이 계산 프로그램 6**

```python
# 원뿔 부피 및 겉넓이 계산 프로그램
# 반지름 과 높이 사용자 입력
rad = int(input("반지름을 입력하세요: "))
hei = int(input("높이를 입력하세요: "))

# 반지름에 음수 넣기를 막음
if rad >= 0:
    # 모선 계산
    sla = pow(rad ** 2 + hei ** 2, 0.5)
    #부피 & 겉넓이 출력
    vol = 1 / 3 * 3.14 * rad ** 2 * hei
    suf = 3.14 * rad ** 2 + 3.14 * rad * sla
    print("원뿔의 부피는", str(vol) + "입니다.")
    print("원뿔의 겉넓이는", str(suf) + "입니다.")
```

**가장 큰 수 찾는 프로그램**

```python
a = int(input("첫 번째 수 입력 : "))
b = int(input("두 번째 수 입력 : "))
c = int(input("세 번째 수 입력 : "))

# 중첩선택구조
# 가장 큰 수 찾기
if a > b:
    if a > c:
        max = a
    else:
        max = c
else:
    if b > c:
        max = b
    else:
        max = c

print("가장 큰 수는", str(max)+"입니다.")
```
